```python
############ using instance, static , class and abstract methods ################
import math
from abc import ABC, abstractmethod

class Pizza(ABC):
    country = 'Italy'
    def __init__(self, radius, ingredients):
        self.radius = radius
        self.ingredients = ingredients

    def __repr__(self):
        return (f'Pizza({self.radius!r}, '
                f'{self.ingredients!r})')

    def area(self):
        print('Instance method :',self.country)
        return self.circle_area(self.radius)

    @staticmethod
    def circle_area(r):
        Pizza.country = 'ITALY'
        print('Static method :', Pizza.country)
        return r ** 2 * math.pi

    @classmethod
    def alternative_contructor(cls):
        # Pizza.country = 'ITALY'
        print('Class method :', Pizza.country)
        return cls(1,[])
    
    @abstractmethod
    def set_flavour(self):
        pass

class CustomPizza(Pizza):
    def set_flavour(self):
        print('Veg pizza')

c = CustomPizza(4, ['mozzarella', 'tomatoes'])
```

```python
# use __slots__ to avoid dynamic addition of attributes for a class
class Car():
    # save RAM by avoiding __dict__
    __slots__ = ['engine','chassis','wheels','body']
    def __init__(self,engine,chassis,wheels,body):
        self.engine = engine
        self.chassis = chassis
        self.wheels = wheels
        self.body = body
    
    def add_attrs(self):
        self.__slots__.append('new_attr')
        self.new_attr = None

c = Car('e','c','w','b')
c.engine = 'e1' # Modifying existing attributes possible
print(c.engine)
c.new_attr = None # Not possible
c.add_attrs() # Not possible
```

```python
## using property decorator to control access to attribute
class Car:
    def __init__(self, color, model, year):
        self.color = color
        self.model = model
        self.year = year
        self._voltage = 12 # making attribute non-public
        #self.voltage = 12 # calls the setter property

    @property
    def voltage(self):
        return self._voltage

    @voltage.setter
    def voltage(self, volts):
        print("Warning: this can cause problems!")
        self._voltage = volts

    @voltage.deleter
    def voltage(self):
        print("Warning: the radio will stop working!")
        del self._voltage
```

```python
########## banking application #############

class Account:
    def __init__(self,name):
        _,self.account_no,self.password = Account.create_credentials()
        self.balance = 0

    @staticmethod
    def create_credentials():
        import random
        import string
        password_set = string.ascii_letters + string.digits + string.punctuation
        acccount_no_set = string.ascii_letters + string.digits
        cif = ''.join([random.choice(string.digits) for _ in range(4)])
        account_no = ''.join([random.choice(acccount_no_set) for _ in range(12)])
        password = ''.join([random.choice(password_set) for _ in range(8)])
        return (cif,account_no,password)

    def credit(self,amount):
        self.balance = self.balance + amount

    def debit(self,amount):
        self.balance = self.balance - amount

    def get_current_balance(self):
        return self.balance


class SavingsBankAccount(Account):
    cif_numbers = {}
    def __init__(self,name,id_proof=None,address_proof=None,interest_rate=4.0):
        Account.__init__(self,name)
        self.id_proof = id_proof
        self.address_proof = address_proof
        self.cif = Account.create_credentials()[0]
        self.interest_rate = interest_rate
        SavingsBankAccount.cif_numbers.update({self.cif:self})

    def get_projected_balance(self,months):
        return (self.balance * months * (self.interest_rate / 1200)) + self.balance


class FixedDepositAccount(Account):
    def __init__(self,cif,name,id_proof,address_proof,cash,interest_rate=6.5):
        if cif in SavingsBankAccount.cif_numbers:
            Account.__init__(self, name)
            sb_cif = cif
        else:
            sb_account = SavingsBankAccount(name,id_proof,address_proof)
            sb_cif = sb_account.cif
            Account.__init__(self, name)
        self.balance = self.balance + cash
        self.interest_rate = interest_rate
        self.linked_sb_account = sb_account
        self.cif = sb_cif

    def get_projected_balance(self,tenure):
        return (self.balance * tenure * (self.interest_rate/1200)) + self.balance

    def get_linked_sb_account(self):
        return self.linked_sb_account

f1 = FixedDepositAccount(cif=None,name='Python',id_proof='Aadhar',address_proof='EB',cash=10000)
f2 = FixedDepositAccount(cif=None,name='Java',id_proof='Aadhar',address_proof='EB',cash=20000)

f1.linked_sb_account.credit(1000)
f2.linked_sb_account.credit(3000)
print(f1.linked_sb_account.__dict__)
print(f2.linked_sb_account.__dict__)
```

```python
# define Python user-defined exceptions
class Error(Exception):
   """Base class for other exceptions"""
   pass

class ValueTooSmallError(Error):
   """Raised when the input value is too small"""
   pass

class ValueTooLargeError(Error):
   """Raised when the input value is too large"""
   pass

# our main program
# user guesses a number until he/she gets it right

# you need to guess this number
number = 10

while True:
   try:
       i_num = int(input("Enter a number: "))
       if i_num < number:
           raise ValueTooSmallError
       elif i_num > number:
           raise ValueTooLargeError
       break
   except ValueTooSmallError:
       print("This value is too small, try again!")
       print()
   except ValueTooLargeError:
       print("This value is too large, try again!")
       print()


print("Congratulations! You guessed it correctly.")
```
